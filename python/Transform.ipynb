```python
#Realizamos un data head
import pandas as pd
import numpy as np

df = pd.read_csv('pax_data_1249_agreements_18-04-20.csv')
df.head()
```

```python
#Mostramos cuales son las variables categóricas
df["Year"]=[df["Dat"][i][0:4] for i in range(len(df["Dat"]))]
cat =['Con','Reg', 'Year','Loc1ISO','Loc2ISO', 'Stage']
num =[ 'GCh', 'GRa', 'GRe', 'HrDem','EqGen', 'DevHum']

total=cat + num
total
```

```python
from sklearn.preprocessing import MinMaxScaler

data_trans=df[total]
data_trans.head(10)

norm_n=[ 'GCh_n', 'GRa_n', 'GRe_n', 'HrDem_n','EqGen_n', 'DevHum_n']
x = data_trans[num]
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
data_trans[norm_n] = pd.DataFrame(x_scaled)

data_trans.shape
```

```python
data_trans
```

```python
data_trans['Con']=[data_trans["Con"][i].replace('(','').replace(')','') for i in range(len(data_trans["Con"]))]
```

```python
import re
con=pd.read_excel('Country.xlsx')
con.head()

a=[]
b=[]
c=[]
d=[]

def todos_elementos_iguales(lista):
    longitud = len(lista)
    for i in range(longitud-1):
        if lista[i] != lista[i+1]:
                return False
    return True



for j in range(len(con["country"])):
    v=[re.search(con["country"][j], data_trans["Con"][i]) for i in range(len(data_trans["Con"]))]
    if todos_elementos_iguales(v)==False:
        for k in range(len(v)):
            if v[k]!=None: 
                a.append(con["country"][j]) 
                b.append(con["code"][j]) 
                c.append(data_trans["Con"][k])                                                         
                d.append(data_trans["Year"][k])
                

peace_con= {"Country": a , 
                  "ISO code": b , 
                  "Con_AGR": c, 
                  "Year": d,
                  }

peacetotal =pd.DataFrame(peace_con)
peacetotal
```
